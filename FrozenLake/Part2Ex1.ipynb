```python
import gymnasium as gym
import matplotlib.pyplot as plt
```

```python
env = gym.make('FrozenLake-v1', is_slippery=True, render_mode='rgb_array')

obs, info = env.reset()

print(f"Initial observation (state): {obs}")
print(f"Additional info: {info}")
```

```python
print(f"Observation space: {env.observation_space}")
print(f"Action space: {env.action_space}")
```

```python
print(f"Previous observation: {obs}")

action = env.action_space.sample() # random action
obs, reward, done, truncated, info = env.step(action)
frame = env.render()

print(f"Action taken: {action}")
print(f"New observation: {obs}")
print(f"Reward recieved: {reward}")
print(f"Episode done? {done}")
print(f"Episode truncated? {truncated}")
print(f"Info: {info}")
```

```python
plt.imshow(frame)
plt.axis('off')
plt.show()
```

```python
obs, info = env.reset()


for i in range(50):
    action = env.action_space.sample() # random action
    obs, reward, done, truncated, info = env.step(action)
    frame = env.render()
    plt.imshow(frame)
    plt.title(f'Step {i+1} Action = {action} Reward = {reward} Done = {done}')
    plt.axis('off')
    plt.show()
    if done or truncated:
        print('Episode ended')
        break
```

```python
env = gym.make('FrozenLake-v1', is_slippery=False, render_mode='rgb_array')

obs, info = env.reset()

def perform_step(i, action: int):
    obs, reward, done, truncated, info = env.step(action)
    frame = env.render()
    plt.imshow(frame)
    plt.title(f'Step {i+1} Action = {action} Reward = {reward} Done = {done}')
    plt.axis('off')
    plt.show()
    return done

actions = [1, 1, 2, 1, 2, 2]
res = False
for i, a in enumerate(actions):
    res = perform_step(i, a)
    if res:
        break
    
```

```python
n = None
best_actions = {0: 1, 1: 2, 2: 1, 3: 0,
               4: 1, 5: n, 6: 1, 7: n,
               8: 2, 9: 1, 10: 1, 11: n,
              12: n, 13: 2, 14: 2, 15: n}

env = gym.make('FrozenLake-v1', is_slippery=False, render_mode='rgb_array')

obs, info = env.reset()

iteration = 0
while True:
    action = best_actions[obs]
    obs, reward, done, truncated, info = env.step(action)
    frame = env.render()
    plt.imshow(frame)
    plt.title(f'Step {iteration+1} Action = {action} Reward = {reward} Done = {done}')
    plt.axis('off')
    plt.show()
    iteration += 1
    if done or truncated:
        print('Episode ended')
        break
```
